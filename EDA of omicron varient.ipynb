```python
world_data=pd.read_csv("/Users/meenashankar/Desktop/dataset/countries of the world.csv")
```

```python
world_data['Country'] = world_data['Country'].apply(lambda x:x.strip())


data = data.merge(world_data[['Country','Region']],left_on='location',right_on='Country')

omicron = data[data['variant']=="Omicron"]
```

```python
data.head()
```

```python
earliest_date = omicron[omicron['num_sequences']>0]['date'].min()
print("First record for Omicron: "+str(earliest_date))
```

```python
tmp = omicron[omicron['date']>=earliest_date][['date','num_sequences']].groupby('date').sum().reset_index()
plt.figure(figsize=(25,5))
plt.plot(tmp['date'],tmp['num_sequences'])
plt.title('Trend for Omicron by Date')
plt.show()
```

```python
tmp = omicron[['location','num_sequences']].groupby('location').sum().reset_index().sort_values(by='num_sequences',ascending=False)
tmp = tmp[tmp['num_sequences']>0]
plt.figure(figsize=(25,5))
plt.bar(tmp['location'],tmp['num_sequences'])
plt.xticks(rotation=90)
plt.title('Distribution of Omicron by countries')
plt.show()
```

```python
tmp = data[['location','variant','num_sequences']].groupby(['location','variant']).sum().reset_index()
tmp_stat = tmp.groupby('location').sum().reset_index()
tmp = tmp.merge(tmp_stat,on="location")
tmp['per_sequences'] = tmp['num_sequences_x']/tmp['num_sequences_y']

pd.pivot_table(tmp,index='location',columns='variant',values='per_sequences').plot(kind='bar', figsize=(25,5), stacked=True)
plt.legend(loc='right')
plt.title('Perception of variant by countries')
plt.show()
```

```python
tmp = omicron[['Region','num_sequences']].groupby('Region').sum().reset_index().sort_values(by='num_sequences',ascending=False)
tmp = tmp[tmp['num_sequences']>0]
plt.figure(figsize=(25,5))
plt.bar(tmp['Region'],tmp['num_sequences'])
plt.xticks(rotation=90)
plt.title('Distribution of Omicron by regions')
plt.show()
```

```python
tmp = data[['Region','variant','num_sequences']].groupby(['Region','variant']).sum().reset_index()
tmp_stat = tmp.groupby('Region').sum().reset_index()
tmp = tmp.merge(tmp_stat,on="Region")
tmp['per_sequences'] = tmp['num_sequences_x']/tmp['num_sequences_y']

pd.pivot_table(tmp,index='Region',columns='variant',values='per_sequences').plot(kind='bar', figsize=(25,5), stacked=True)
plt.legend(loc='right')
plt.title('Perception of variant by regions')
plt.show()
```
